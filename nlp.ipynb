```python
import gradio as gr
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

def sentiment_analysis(text):

    result = classifier(text)
    return result[0]['label'], result[0]['score']


iface = gr.Interface(fn=sentiment_analysis,
                     inputs=gr.Textbox(label="Enter Text for Sentiment Analysis"),
                     outputs=[gr.Textbox(label="Sentiment"), gr.Textbox(label="Confidence Score")])


iface.launch()
```

```python
import spacy
import gradio as gr


nlp = spacy.load("en_core_web_sm")

def pos_tagging(text):
    doc = nlp(text)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags


iface = gr.Interface(fn=pos_tagging,
                     inputs=gr.Textbox(label="Enter Text for POS Tagging"),
                     outputs=gr.Dataframe(headers=["Word", "POS Tag"], type="array"))


iface.launch()
```

```python
import gradio as gr
from googletrans import Translator


translator = Translator()


def translate_text(text, target_language='en'):

    translation = translator.translate(text, dest=target_language)
    return translation.text


iface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text to Translate"),
        gr.Dropdown(
            label="Select Target Language",
            choices=["en", "es", "fr", "de", "it", "pt", "ja", "ko", "zh-cn"],
            value="en"
        ),
    ],
    outputs=gr.Textbox(label="Translated Text"),
)

iface.launch()
```

```python
import gradio as gr
from transformers import MarianMTModel, MarianTokenizer


model_name = 'Helsinki-NLP/opus-mt-en-es'
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)


def translate_text(text, target_language='en'):

    if target_language == 'es':
        model_name = 'Helsinki-NLP/opus-mt-en-es'
    elif target_language == 'fr':
        model_name = 'Helsinki-NLP/opus-mt-en-fr'
    else:
        model_name = 'Helsinki-NLP/opus-mt-en-en'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)


    tokens = tokenizer.encode(text, return_tensors="pt")
    translated = model.generate(tokens, max_length=100)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text


iface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Enter Text to Translate"),
        gr.Dropdown(
            label="Select Target Language",
            choices=["en", "es", "fr", "de", "it", "pt", "ja", "ko", "zh-cn"],
            value="en"
        ),
    ],
    outputs=gr.Textbox(label="Translated Text"),
)


iface.launch()
```

```python
import gradio as gr
from transformers import pipeline

summarizer = pipeline("summarization")


def summarize_text(text):
    summary = summarizer(text, max_length=150, min_length=30, do_sample=False)
    return summary[0]['summary_text']

iface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(label="Enter Text to Summarize", lines=6, placeholder="Enter your paragraph here..."),
    outputs=gr.Textbox(label="Summarized Text"),
    title="Text Summarization",
    description="This app summarizes your paragraph into a concise form."
)


iface.launch()
```

```python
import gradio as gr
from transformers import pipeline


qa_pipeline = pipeline("question-answering")


def answer_question(question, context):
    answer = qa_pipeline(question=question, context=context)
    return answer['answer']

iface = gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Textbox(label="Enter Context (Paragraph or Article)", lines=6, placeholder="Enter context here..."),
        gr.Textbox(label="Ask a Question", placeholder="Ask your question here..."),
    ],
    outputs=gr.Textbox(label="Answer"),
    title="Question-Answering Application",
    description="This app answers your question based on the provided context."
)


iface.launch()
```

```python
import gradio as gr
import language_tool_python


tool = language_tool_python.LanguageTool('en-US')

def grammar_and_spell_check(text):
    matches = tool.check(text)
    corrected_text = language_tool_python.utils.correct(text, matches)


    errors = [(match.context, match.message) for match in matches]

    return corrected_text, errors


iface = gr.Interface(
    fn=grammar_and_spell_check,
    inputs=gr.Textbox(label="Enter Text", lines=6, placeholder="Type your text here..."),
    outputs=[gr.Textbox(label="Corrected Text"), gr.Dataframe(label="Detected Errors", headers=["Context", "Error Message"], type="array")],
    title="Grammar and Spell Checker",
    description="This app checks grammar and spelling in your text and suggests corrections."
)


iface.launch()
```

```python
import gradio as gr
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


documents = [
    "Information retrieval is the process of obtaining relevant information from a collection of documents.",
    "Search engines like Google use advanced algorithms for information retrieval.",
    "Machine learning is transforming the way information retrieval systems work.",
    "The study of algorithms in the context of information retrieval is very important.",
    "Natural Language Processing is a subset of machine learning and is crucial for information retrieval tasks."
]


vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

def retrieve_documents(query):

    query_tfidf = vectorizer.transform([query])

    cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix)

    ranked_indices = cosine_similarities[0].argsort()[::-1]


    top_documents = [documents[i] for i in ranked_indices[:3]]
    return top_documents

iface = gr.Interface(
    fn=retrieve_documents,
    inputs=gr.Textbox(label="Enter your Query", placeholder="Type your query here..."),
    outputs=gr.Textbox(label="Top 3 Relevant Documents", lines=5),
    title="Information Retrieval System",
    description="This is a simple Information Retrieval system that ranks documents based on the relevance to your query."
)


iface.launch()
```
